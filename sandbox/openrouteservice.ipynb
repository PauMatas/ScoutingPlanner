```python
import requests
import json
```

```python
with open('../etc/config.json') as json_file:
    config = json.load(json_file)

api_key = config['openrouteservice']['api_key']
```

```python


# Definim les coordenades de sortida i arribada
lat1 = 41.5368
lon1 = 1.6626
lat2 = 41.3851
lon2 = 2.1734

# Fem la petició a la API
url = f'https://api.openrouteservice.org/v2/directions/driving-car?api_key={api_key}&start={lon1},{lat1}&end={lon2},{lat2}'
response = requests.get(url)

# Convertim la resposta a un diccionari de Python
data = json.loads(response.text)

# Obtenim les dades de la ruta
duration = data['features'][0]['properties']['segments'][0]['duration']
distance = data['features'][0]['properties']['segments'][0]['distance']

# Mostrem els resultats
print(f'Durada de la ruta: {duration/60} minuts')
print(f'Distància de la ruta: {distance/1000} km')
```
